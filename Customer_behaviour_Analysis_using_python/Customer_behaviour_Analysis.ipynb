```python
import pyodbc
from sqlalchemy import create_engine
import pandas as pd
import urllib

# ── 1. Load CSV ───────────────────────────────────────────────
df = pd.read_csv(r"D:\RESUME_STUFF\Customer_behaviour_Project\customer_shopping_behavior.csv")
print(f"Total rows loaded: {len(df)}")        # e.g. 3897 rows
print(f"Total columns: {len(df.columns)}")    # e.g. 18 columns

# ── 2. Connect to SQL Server ──────────────────────────────────
server = r'HP\SQLEXPRESS'
database = 'customer_behaviour_Analysis'

connection_string = (
    f"DRIVER={{ODBC Driver 17 for SQL Server}};"
    f"SERVER={server};"
    f"DATABASE={database};"
    f"Trusted_Connection=yes;"
)

params = urllib.parse.quote_plus(connection_string)
engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")
print("✅ Connected to SQL Server!")

# ── 3. Push ALL rows to SQL Server ───────────────────────────
print("Pushing data... please wait ⏳")

df.to_sql(
    name='customer_shopping_behavior',
    con=engine,
    if_exists='replace',   # Drops old table & recreates fresh
    index=False,           # Don't add extra index column
    chunksize=500          # Sends 500 rows per batch (all rows still go in)
)

print(f"✅ ALL {len(df)} rows successfully pushed!")

# ── 4. Verify row count in SQL Server ────────────────────────
result = pd.read_sql("SELECT COUNT(*) AS total_rows FROM customer", engine)
print(f"✅ Rows confirmed in SQL Server: {result['total_rows'][0]}")
```
